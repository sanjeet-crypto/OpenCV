```python
import cv2
import numpy as np
import time

cap = cv2.VideoCapture('vid.mp4')

start= time.time()
Y_start=0


if(cap.isOpened()==False):
    print("Error opening video Stream of File")
    
while(cap.isOpened()):
    ret,frame=cap.read()
    if ret==True:
        
        duplicate_img = frame #make a copy of the original image
        
        black_img=np.zeros((frame.shape[0],frame.shape[1], 3))
        
         
        grey=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        #kernel = np.ones((30, 30), np.uint8)
        
        canny_edg=cv2.Canny(grey,30,180)
        #cv2.imshow('canny edge detection',canny_edg)
        #if cv2.waitKey(25) == ord('q'):
            #break
        
        dup = canny_edg.copy()
        
        contours,hierarchy = cv2.findContours(dup, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        
        for cnt in contours:
            area=cv2.contourArea(cnt)
           
            if area>3000:
                
                cent_moment = cv2.moments(cnt)
                
                stop=time.time()
                diff=start-stop
                
                print(diff)
                
                X = int(cent_moment["m10"] / cent_moment["m00"])
                Y = int(cent_moment["m01"] / cent_moment["m00"])
                
               
                cv2.circle(black_img, (X,Y), 8, (255,255,255), -1)
                
                velocity= (Y-Y_start)/diff
                Y_start= Y
                start=stop
                
                print(velocity)
                if velocity<0:
                    cv2.arrowedLine(black_img, ((X-20),Y), (X-20,Y-20),(0,255,255), 4,8,0,0.3)
                if velocity>0:
                    cv2.arrowedLine(black_img, (X-20,Y-20),((X-20),Y) ,(0,255,255), 4,8,0,0.3)
                if velocity==0:
                    continue
                              
                                
                cv2.putText(black_img," ("+str(X)+","+str(Y)+")",(X-15,Y-15),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),1)
                cv2.putText(black_img,str(velocity)+"pixels/sec",(X-30,(Y-40)),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,255),2)
        
        
        
        cv2.imshow("Contour with centroids",black_img )
        if cv2.waitKey(25) == ord('q'):
            break
    
            
               
    else:
        break
        
cap.release()
cv2.destroyAllWindows()
        
        
            
```
